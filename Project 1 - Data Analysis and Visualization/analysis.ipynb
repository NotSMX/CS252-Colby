### Analysis and summary statistic methods

To actually perform computations on data, you will implement the following methods in `analysis.py`, modeled off of common functions in the `pandas` python package (https://pandas.pydata.org/pandas-docs/stable/user_guide/computation.html) that we will adopt later in the semester:

- `set_data(data)`
- `min(headers, rows=[])`: Computes the minimum of each variable in `headers` in the data object. Possibly only in a subset of data samples (`rows`) if `rows` is not empty.
- `max(headers, rows=[])`: Computes the maximum of each variable in `headers` in the data object.
- `range(headers, rows=[])`: Computes the range [min, max] for each variable in `headers` in the data object.
- `mean(headers, rows=[])`: Computes the mean for each variable in `headers` in the data object. Equation for mean: $$\bar{x} = \frac{1}{N} \sum_{i=1}^N x_i$$ where $x_i$ is the $i^{th}$ data sample and $N$ is the total number of data samples in the dataset.
- `var(headers, rows=[])`: Computes the variance for each variable in `headers` in the data object. Equation for sample variance: $$s^2 = \frac{1}{N-1} \sum_{i=1}^N (x_i - \bar{x})^2$$ where $x_i$ is the $i^{th}$ data sample, $N$ is the total number of data samples in the dataset, $\bar{x}$ is the mean.
- `std(headers, rows=[])`: Computes the standard deviation for each variable in `headers` in the data object. Equation for sample standard deviation: $$s^2 = \sqrt{\frac{1}{N-1} \sum_{i=1}^N (x_i - \bar{x})^2}$$ where $x_i$ is the $i^{th}$ data sample, $N$ is the total number of data samples in the dataset, $\bar{x}$ is the mean.

```python
samplestring = '09/17/1908,17:18,"Fort Myer, Virginia",Military - U.S. Army,,Demonstration,Wright Flyer III,,1,2,1,0,"During a demonstration flight, a U.S. Army flyer flown by Orville Wright nose-dived into the ground from a height of approximately 75 feet, killing Lt. Thomas E. Selfridge who was a passenger. This was the first recorded airplane fatality in history.  One of two propellers separated in flight, tearing loose the wires bracing the rudder and causing the loss of control of the aircraft.  Orville Wright suffered broken ribs, pelvis and a leg.  Selfridge suffered a crushed skull and died a short time later."'
print(samplestring.strip(','))
```

Entries like "Fort Myer, Virginia" became 2 different entries despite being one location. So in order to troubleshoot this without adding more modules,(because I honestly had no idea what I was looking at online) I decided to make spliting its own method, which code is listed below.

```python
 def split(self, line):
        uglyLine = line.split('"')
        result = []
        for d in range(0, len(uglyLine)) :
            if uglyLine[d] != '\n': 
                # odd numbers are the ones in quotes
                if d % 2 == 0 :
                    splitline = uglyLine[d].split(',')
                    for a in range(0,len(splitline)) :
                        thing = splitline[a]
                        if thing == '' :
                            if a == 0 or a == len(splitline) - 1 :
                                pass
                            else:  
                                result.append(thing)
                        else:
                            result.append(thing)
                else :
                    result.append(uglyLine[d])
        
        return result
```

Luckily, the sentences that did have ignorable commas were put in quotations, so I decided to utilize that.

<b>The basic premise: for each line in the CSV file, this method is called. It splits it based on quotations (result is known as uglyline), and, based on the index of this "uglyLine", chooses whether or not to ignore splitting by commas. (SPOILERS, IT'S WHETHER THE INDEX IS EVEN OR ODD)<b>

The reason why this works is that each time a line is split by a quotation mark, it moves up an index, and moves up again when it ends. So if we were to start reading a line, no matter where, a quotation mark makes uglyLine[0] and uglyLine[1], with the latter being what's inside the quotation mark. We know where it ends based on uglyLine[2], because that point is where another quote appears. The next one would be uglyLine[3], and so on.

<b>In summary, if the index of uglyline is odd, don't split by commas.<b>

We can then determine when to split by commas by writing a simple if statement.



Unfortunately, it wasn't that simple. There are a couple issues that come up when trying to perform this, which I'll go over here.

<b>Problem 1. Splitting twice makes a list within a list, which we don't want.<b>

I had to make a temp list called result that would end up being returned, and in the case where I had to split twice (for commas) I would have to go back inside and append each of them individually to the result list. For things split by quotes, since whatever's in quotes is only one entry, I only needed to append that as well.

<b>Problem 2. Splitting by quotes also leaves trailing commas, which when split by them makes unwanted extra headers.<b>

If a line for example was 'thing, "another, thing", thingy' and split by quotes, it would become ['thing,', 'another, thing', ', thingy'] 

Then, if we were to split by commas, it would become ['thing', '', 'another, thing', '', ' thingy'] which adds 2 extra entries due to the commas before and after the quotes. Lucikly, we can easily check this with an if statement (== '').

We inadvertedly make another problem: since there are some entries within this dataset that have missing values, we may accidentally exclude them from being a part of the line and mess up our numbers again. Thankfully I realized this problem only occurs if the space is the last or first within the splitline, so we just added another if statement.

i.e. ['thing,,,'] ==> ['thing', '', ''] (last one is removed) ==> ['thing', ''] (good!)

<b>Problem 3. Scenarios where the quotation mark is the last character in the line<b>

np.nans also mess with the plotting a little, so I needed to make the code so I could ignore it.
